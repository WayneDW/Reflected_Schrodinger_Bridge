```python
import numpy as np
```

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
import scipy.stats
from tqdm import tqdm
```

```python
from tools import HelperTorch, Sampler
from domain import Flower, Polygon, Heart, Cross, Star
```

```python
np.set_printoptions(precision=3)
np.set_printoptions(suppress=True)
np.random.seed(9)
```

```python
def plot_domain(myDomainClass):
    points = myDomainClass.position(np.arange(0, 1, 0.001))
    fig = plt.figure(figsize=(3, 3))
    plt.plot(points[0, :], points[1, :])
    plt.legend(loc="upper left", prop={'size': 10})
    plt.tight_layout()
    plt.show()
```

```python
def plot_MCMC_traj(myHelper, iters=1e5, lr=1e-4, boundary_=1.5):
    sampler = Sampler(myHelper, boundary=boundary_, xinit=[0.,0.], lr=lr, T=2)
    for iters in tqdm(range(int(iters))):
        sampler.rgld_step(iters)
        if iters % 50000 == 50000-1:
            fig = plt.figure(figsize=(4, 4))
            plt.plot(sampler.list[0, :], sampler.list[1, :], \
                     linewidth=0.3, marker='.', markersize=3, color='r', label="Iteration="+str(iters)) 
            plt.legend(loc="upper left", prop={'size': 10})
            plt.xlim([-boundary_, boundary_])
            plt.ylim([-boundary_, boundary_])
            plt.tight_layout()
            plt.show()
```

```python
myFlower = Flower(petals=5, move_out=3)
plot_domain(myFlower)
```

```python
myHelper = HelperTorch(Flower, grid_radius=1e-2, grid_curve=1e-3)
```

```python
plot_MCMC_traj(myHelper, iters=1e6, lr=3e-5, boundary_=1.5)
```

```python
myHeart = Heart()
plot_domain(myHeart)
```

```python
myHelper = HelperTorch(Heart, grid_radius=1e-2, grid_curve=1e-3)
```

```python
plot_MCMC_traj(myHelper, iters=1e6, lr=3e-5, boundary_=2)
```

```python
myHelper = HelperTorch(Polygon)
plot_MCMC_traj(myHelper, iters=3e5, lr=3e-5, boundary_=1.5)
```
